Для сравнения каждого из законов воспользуемся графическим методом, чтобы по графикам сказать какой лучше:

```python
import matplotlib.pyplot as plt
import numpy as np
import math

x_points = np.array([1.09, 1.41, 1.78, 2.19, 2.63, 3.16, 3.72, 4.16, 4.53, 5.22])
y_points = np.array([-5.0,-4.22, -3.44, -2.81, -2.22, -1.63, -1.13, -0.75, -0.47, 0.0])

f1 = lambda x: 1.17240363343436 * x - 5.67131446033531

f2 = lambda x: -0.202523795964537 * x ** 2 + 2.42513202824543 * x - 7.24959599411749

f3 = lambda x: 0.754711013199889 - 6.85320490739780 / x

f4 = lambda x: -5.29675293078252 + 3.19027361229652 * math.log(x)

array_x = np.linspace(1, 6, 1000)

y1 = [f1(x_i) for x_i in array_x]
y2 = [f2(x_i) for x_i in array_x]
y3 = [f3(x_i) for x_i in array_x]
y4 = [f4(x_i) for x_i in array_x]

plt.scatter(x_points, y_points, color='red', s=60, zorder=5, label="Исходные точки")

plt.plot(array_x, y1, label="y=ax+b", color='yellow', lw=2)

plt.plot(array_x, y2, label="y=ax^2+bx+c", color='green', ls='--')

plt.plot(array_x, y3, label="y=a+b/x", color='blue', ls='--')

plt.plot(array_x, y4, label="y=a+b*ln(x)", color='black', ls='--')

plt.title("Сравнение законов", fontsize=14, weight='bold', pad=15)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10, loc="best", frameon=True)
plt.tight_layout()
plt.show()
```
