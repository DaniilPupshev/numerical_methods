Выбраны следующие законы:
1. Линейный (y=ax+b)
2. Квадратичный (y=ax^2+bx+c)
3. Гиперболический (y=a+(b/x))
4. Логарифмический (y=a+bln(x))

------------

Произведем промежуточные вычисления для закона 1:

```python
import sympy as sp

X = [1.09, 1.41, 1.78, 2.19, 2.63, 3.16, 3.72, 4.16, 4.53, 5.22]
Y = [-5.0,-4.22, -3.44, -2.81, -2.22, -1.63, -1.13, -0.75, -0.47, 0.0]

F = 0

a = sp.Symbol('a')
b = sp.Symbol('b')

for i in range(len(X)):
    F += ((a * X[i] + b - Y[i]) ** 2) # Закон под номером 1
print(f'Промежуточный рассчет:\n{sp.expand(F)}')
```

```python
# ==== Найдем производные относительно каждой неизвестной ====
f = 106.9565*a**2 + 59.78*a*b + 88.2388*a + 10*b**2 + 43.34*b + 72.1837

der_a = sp.diff(f, a)
der_b = sp.diff(f, b)

print(f'Итого, имеем систему:\n/\n| {der_a} = 0\n< {der_b} = 0\n|\n\\')
```

Решим систему методом Гаусса и найдем a, b, c:

```python
from sympy import pprint, Matrix

A_B = Matrix([
    [213.913, 59.78, -88.2388],
    [59.78, 20.0, -43.34],
])

step_matrix = A_B.rref()[0]
pprint(step_matrix)

answer = []
for i in range(2, step_matrix.shape[0] * step_matrix.shape[1], 3):
    answer.append(step_matrix[i])

a, b = answer

# Выводим ответ
print(f'\nДля закона номер 1:\na = {a}\nb = {b}')
print(f'\n\nИтого, вид закона: y = {a}x + ({b})')
```
