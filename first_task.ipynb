Для начала определим функцию, с которой будем работать и зададим границы интегрирования:

```python
from math import sin, cos

a = 0.3
b = 1.3
p = 4
n = int(input())
f = lambda x: (sin(0.4 * x + 0.9)) / (2.1 + cos(4 * (x ** 2) + 0.2))
```

Далее находим шаг интегрирования используя формулы ниже (т.к. n задано по условию):

1) 

```python
# ==== Находим h ====
h = abs(b - a) / n

print(f'h = {round(h, 4)}')
```

Нахождение интеграла по формуле Симпсона:

2) 
   
3) 

```python
def make_y(f, a, n, h):
    xs = [a + i * h for i in range(n + 1)]
    ys = [f(x) for x in xs]
    return ys

# Формула Симпсона
def I(h, y_array):
    n = len(y_array) - 1
    odd_sum  = sum(y_array[1:n:2])
    even_sum = sum(y_array[2:n:2])
    return (h / 3) * (y_array[0] + y_array[n] + 4 * odd_sum + 2 * even_sum)

y_n_array = make_y(f, a, n, h)
I_c_h = I(h, y_n_array)

# Пересчитываем сетку для шага 2h
n2 = n // 2
h2 = 2 * h
y_n2_array = make_y(f, a, n2, h2)
I_c_2h = I(h2, y_n2_array)

# Погрешность
fallibility = abs(I_c_h - I_c_2h) / (2 ** p - 1)

print(f'I(h) = {round(I_c_h, 4)}\nI(2h) = {round(I_c_2h, 4)}')
print(f'Погрешность = {round(fallibility, 4)}\n\n')
print(f'Итого, решение интеграла: {round(I_c_h, 4)} +- {round(fallibility, 4)}')
```
