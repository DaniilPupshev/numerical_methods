Найдем полином Ньютона:

```python
import sympy as sp

# Символ переменной
x = sp.Symbol('x')

# Узлы (x_i)
x0, x1, x2, x3 = 0.079, 0.637, 1.345, 2.095

# Коэффициенты Ньютона (из таблицы разделённых разностей)
f0 = -4.308
f1 = 6.39605734767025
f2 = -2.3344228922342456
f3 = 1.1264175385584168
f4 = -0.45906450186634273

# Формула полинома Ньютона
N = (
    f0
    + f1 * (x - x0)
    + f2 * (x - x0) * (x - x1)
    + f3 * (x - x0) * (x - x1) * (x - x2)
    + f4 * (x - x0) * (x - x1) * (x - x2) * (x - x3)
)

N_expanded = sp.expand(N)

print(f'Полином Ньютона имеет вид:\ny={N_expanded}')
```

```python
f = lambda x: -0.459064501866343 * x ** 4 + 3.03428960831494 * x ** 3 - 7.10330564955387 * x ** 2 + 10.2145956177061 * x - 5.07209946575458
print(f'N_4(x1+x2)={f(x1 + x2)}')
```

Построим график полинома:

```python
import matplotlib.pyplot as plt
import numpy as np

# ==== Строим график полинома Ньютона ====
array_x = np.linspace(0, 3, 1000)
y = [f(x_i) for x_i in array_x]

plt.plot(array_x, y, label="Полином Ньютона: f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
