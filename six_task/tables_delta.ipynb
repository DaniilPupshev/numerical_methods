```python
X = [0.079, 0.637, 1.345, 2.095, 2.782]
Y = [-4.308, -0.739, 1.697, 4.208, 6.203]
```

Сформируем таблицу конечных разностей:

```python
k = len(X) - 1 # Берем на единицу меньше, т.к. отсчет с 0

table_finite_diff = [
    [i for i in range(len(X))], # k
    X, # X_k
    Y, # Y_k (далее будут добавляться delta_n_Y_k)
]

while k >= 1:
    tmp = [] # Векторно заданный k-тый столбец дельты таблицы
    for y_k in range(k):
        tmp.append(table_finite_diff[-1][y_k + 1] - table_finite_diff[-1][y_k])
    table_finite_diff.append(tmp)
    k -= 1 # Переходим к генерации следующего столбца

print(f'Таблица конечных разностей в векторном виде:\n{table_finite_diff}')
```

Сформируем таблицу разделенных разностей, используя общую формулу вида:

1. 

```python
k = len(X) - 1 # Обновляем порядок таблицы

# Обновляем таблицу
table_div_diff = [
    [i for i in range(len(X))], # k
    X, # X_k
    Y, # Y_k (далее будут добавляться k-тый порядок)
]

prev = Y[:]

while k >= 1:
    tmp = [] # Векторно заданный k-тый столбец
    for i in range(k):
        dx = X[i + (len(X) - k)] - X[i] # Считаем знаменатель
        if dx != 0:
            tmp.append((prev[i + 1] - prev[i]) / dx)
    table_div_diff.append(tmp) # Добавляем новый созданный вектор
    prev = tmp
    k -= 1 # Переходим к генерации следующего столбца
    
print(f'Таблица разделенных разностей в векторном виде:\n{table_div_diff}')
```
