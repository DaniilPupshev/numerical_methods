```python
X = [0.079, 0.637, 1.345, 2.095, 2.782]
Y = [-4.308, -0.739, 1.697, 4.208, 6.203]
```

   :

```python
k = len(X) - 1 #    , ..   0

table_finite_diff = [
    [i for i in range(len(X))], # k
    X, # X_k
    Y, # Y_k (   delta_n_Y_k)
]

while k >= 1:
    tmp = [] #   k-   
    for y_k in range(k):
        tmp.append(table_finite_diff[-1][y_k + 1] - table_finite_diff[-1][y_k])
    table_finite_diff.append(tmp)
    k -= 1 #     

print(f'     :\n{table_finite_diff}')
```

```python
k = len(X) - 1 #   

#  
table_div_diff = [
    [i for i in range(len(X))], # k
    X, # X_k
    Y, # Y_k (   k- )
]

prev = Y[:]

while k >= 1:
    tmp = [] #   k- 
    for i in range(k):
        dx = X[i + (len(X) - k)] - X[i] #  
        if dx != 0:
            tmp.append((prev[i + 1] - prev[i]) / dx)
    table_div_diff.append(tmp) #    
    prev = tmp
    k -= 1 #     
    
print(f'     :\n{table_div_diff}')
```
