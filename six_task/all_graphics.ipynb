Возьмем все графики и объединими их в одной плоскости:

```python
import matplotlib.pyplot as plt
import numpy as np

# ===== Определяем функции =====
L = lambda x: -0.459064501866344 * x ** 4 + 3.03428960831494 * x ** 3 - 7.10330564955388 * x ** 2 + 10.2145956177061 * x - 5.07209946575458
N = lambda x: -0.459064501866343 * x ** 4 + 3.03428960831494 * x ** 3 - 7.10330564955387 * x ** 2 + 10.2145956177061 * x - 5.07209946575458

s1 = lambda x: 6.396 * x - 4.813
s2 = lambda x: 3.441 * x - 2.931
s3 = lambda x: 3.348 * x - 2.806
s4 = lambda x: 2.904 * x - 1.876

s1_q = lambda x: -2.334 * x ** 2 + 8.068 * x - 4.931
s2_q = lambda x: -0.309 * x ** 2 + 4.411 * x - 3.677

# ===== Диапазоны =====
array_x = np.linspace(0, 3, 1000)
array_x1_s = np.linspace(0, 0.637, 200)
array_x2_s = np.linspace(0.637, 1.345, 200)
array_x3_s = np.linspace(1.345, 2.095, 200)
array_x4_s = np.linspace(2.095, 2.782, 200)
array_x1_sq = np.linspace(0, 1.345, 400)
array_x2_sq = np.linspace(1.345, 2.782, 400)

# ===== Значения =====
y_L = [L(x_i) for x_i in array_x]
y_N = [N(x_i) for x_i in array_x]
y1_s = [s1(x_i) for x_i in array_x1_s]
y2_s = [s2(x_i) for x_i in array_x2_s]
y3_s = [s3(x_i) for x_i in array_x3_s]
y4_s = [s4(x_i) for x_i in array_x4_s]
y1_sq = [s1_q(x_i) for x_i in array_x1_sq]
y2_sq = [s2_q(x_i) for x_i in array_x2_sq]

# ===== График =====
plt.figure(figsize=(10, 6), facecolor="#fafafa")

# Полиномы
plt.plot(array_x, y_L, label="Полином Лагранжа", color='gold', lw=3.5)
plt.plot(array_x, y_N, label="Полином Ньютона", color='dodgerblue', lw=2.5, ls='--')

# Линейный сплайн (подпись только у первой линии)
plt.plot(array_x1_s, y1_s, color='red', lw=2, ls='-.', label="Линейный сплайн")
plt.plot(array_x2_s, y2_s, color='red', lw=2, ls='-.')
plt.plot(array_x3_s, y3_s, color='red', lw=2, ls='-.')
plt.plot(array_x4_s, y4_s, color='red', lw=2, ls='-.')

# Квадратичный сплайн
plt.plot(array_x1_sq, y1_sq, color='green', lw=2.5, label="Квадратичный сплайн", ls=':')
plt.plot(array_x2_sq, y2_sq, color='green', lw=2.5, ls=':')

# ===== Оформление =====
plt.title("Сравнение интерполяционных функций", fontsize=14, weight='bold', pad=15)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10, loc="best", frameon=True)
plt.tight_layout()
plt.show()
```
