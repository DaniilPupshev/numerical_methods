В данном файле с помощью программного кода будет получено локализованное значение корня уравнения (решение пункта 1).
Для этого мы прировняем две функции и построим графики (точка пересечения будет лежать в нужном нам промежутке):

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.optimize import fsolve

f = lambda x: math.e ** (-.5 * x) - .2 * (x ** 2) + 1
f1 = lambda x: math.e ** (-.5 * x) + math.e ** 0
f2 = lambda x: .2 * (x ** 2)

# ==== Создадем массивы значений для построения графиков ====
array_x = np.linspace(-5, 5, 1000)

array_y1 = [f1(x) for x in array_x]
array_y2 = [f2(x) for x in array_x]

# ==== Находим точку пересечения и строим графики ====
x_intersection, y_intersection = 0, 0

for x in array_x:
    x_intersection = fsolve(f, x)[0]
y_intersection = f1(x_intersection)
print(f'Точка пересечения:\nX = {x_intersection}\nY = {y_intersection}')

plt.plot(array_x, array_y1, label="f1(x)=e^(-0.5x)+1")
plt.plot(array_x, array_y2, label="f2(x)=0.2*(x^2)")
plt.plot(x_intersection, y_intersection, "ro", markersize=8, label="Точка пересечения")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ==== Локализуем корень уравнения ====
# Пусть левая граница будет равна a, парвая - b

a = int(x_intersection)
b = int(x_intersection) + 1
print(f'Итого, корень уравнения находится в промежутке:\nx ∈ [{a};{b}]')
```
