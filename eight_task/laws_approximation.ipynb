```python
import sympy as sp

X = [0.079, 0.637, 1.345, 2.095, 2.782]
Y = [-4.308, -0.739, 1.697, 4.208, 6.203]

F = 0

a = sp.Symbol('a')
b = sp.Symbol('b')
c = sp.Symbol('c')

for i in range(len(X)):
    F += ((a * X[i] ** 2 + b * X[i] + c - Y[i]) ** 2) #    1
print(f' :\n{sp.expand(F)}')
```

```python
# ====      ====
f = 82.601031079268*a**2 + 66.83693932*a*b + 28.699168*a*c - 138.440900956*a + 14.349584*b**2 + 13.876*b*c - 55.087792*b + 5*c**2 - 14.122*c + 78.169267

der_a = sp.diff(f, a)
der_b = sp.diff(f, b)
der_c = sp.diff(f, c)

print(f',  :\n/\n| {der_a} = 0\n< {der_b} = 0\n| {der_c} = 0\n\\')
```

```python
step_matrix = A_B.rref()[0]
pprint(step_matrix)
```

```python
answer = []
for i in range(3, step_matrix.shape[0] * step_matrix.shape[1], 4):
    answer.append(step_matrix[i])

a, b, c = answer

#  
print(f'   1:\na = {a}\nb = {b}\nc = {c}')
```
