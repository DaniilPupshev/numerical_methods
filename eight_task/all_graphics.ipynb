Построим графики всех функций:

```python
import matplotlib.pyplot as plt
import numpy as np
import math

x_points = np.array([0.079, 0.637, 1.345, 2.095, 2.782])
y_points = np.array([-4.308, -0.739, 1.697, 4.208, 6.203])

f = lambda x: -0.646365589559600 * x ** 2 + 5.60655995513196 * x - 4.51244712932210

f2 = lambda x: 2.27137531819158 * x -4.86167846235398 * (1 / math.e ** x) + 0.106361656693819

array_x = np.linspace(0, 3, 1000)

y1 = [f(x_i) for x_i in array_x]
y2 = [f2(x_i) for x_i in array_x]

plt.scatter(x_points, y_points, color='red', s=60, zorder=5, label="Исходные точки")

plt.plot(x_points, y_points, label="График исходной функции")

plt.plot(array_x, y1, label="y=ax^2+bx+c", color='yellow', lw=2)

plt.plot(array_x, y2, label="y=ax+b*e^(-x)+c", color='green', ls='--')

plt.title("Сравнение законов", fontsize=14, weight='bold', pad=15)
plt.xlabel("x", fontsize=12)
plt.ylabel("y", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=10, loc="best", frameon=True)
plt.tight_layout()
plt.show()
```
