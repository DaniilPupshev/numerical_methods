```python
import sympy as sp
import math

E = 10 ** (-6)
x0 = 3
x = sp.symbols('x')

f = lambda x: math.e ** (-.5 * x) - .2 * (x ** 2) + 1
f_derivative1 = sp.diff(math.e ** (-.5 * x) - .2 * (x ** 2) + 1, x) #   - f'(x)
f_derivative2 = sp.diff(math.e ** (-.5 * x) - .2 * (x ** 2) + 1, x, 2) #   - f''(x)
print(f"f'(x)={f_derivative1}\nf``(x)={f_derivative2}")
```
