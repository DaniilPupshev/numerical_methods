```python
import sympy as sp
import math

E = 10 ** (-6)
x0 = 2
x1 = 3
x = sp.symbols('x')

f = lambda x: math.e ** (-.5 * x) - .2 * (x ** 2) + 1
```

      : xn+1 = xn-1 - ((f(xn-1)) / (f(xn) - f(xn-1))) * (xn - xn-1)

```python
max_iter = 100
x_prev = x0
x = x1

for i in range(max_iter):
    f_prev, f_curr = f(x_prev), f(x)
    denom = f_curr - f_prev
    x_next = x_prev - f_prev * (x - x_prev) / denom
    if abs(x_next - x) < E:
        print(f" : x  {round(x_next, 6)}\n: {i+1}")
        break
    x_prev, x = x, x_next
```
