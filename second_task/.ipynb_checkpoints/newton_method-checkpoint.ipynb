В этом файле уравнение будет решено методом Ньютона (пункт 3). Для начала выведем f'(x) и f''(x):

```python
import sympy as sp
import math

E = 10 ** (-6)
x0 = 3
x = sp.symbols('x')

f = lambda x: math.e ** (-.5 * x) - .2 * (x ** 2) + 1
f_derivative1 = sp.diff(math.e ** (-.5 * x) - .2 * (x ** 2) + 1, x) # первая производная - f'(x)
f_derivative2 = sp.diff(math.e ** (-.5 * x) - .2 * (x ** 2) + 1, x, 2) # вторая производная - f''(x)
print(f"f'(x)={f_derivative1}\nf``(x)={f_derivative2}")
```

Проверяем условие сходимости: f(x0) * f''(x0) > 0

```python
f_d1 = lambda x: -0.4 * x - (0.5 / (math.e ** (0.5 * x)))
f_d2 = lambda x: -0.4 * x + (0.25 / (math.e ** (0.5 * x)))

if f(x0) * f_d2(x0) > 0:
    print(f'f(x0)={f(x0)}\nf``(x0)={f_d2(x0)}\nУсловие выполнено!')
```

С помошью интерационного процесса находим корень уравнения методом Ньютона. Итерационный процесс имеет вид: xn+1 = xn - ((f(xn)) / (f'(xn)))

```python
max_iter = 100
xn = x0

for i in range(max_iter):
    x_next = xn - ((f(xn)) / (f_d1(xn)))
    if abs(x_next - xn) <= E:
        print(f"Решение найдено: x ≈ {round(x_next, 6)}\nИтераций: {i+1}")
        break
    xn = x_next
```
