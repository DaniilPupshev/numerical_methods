На основании системы составим матрицу:

```python
from sympy import symbols, Matrix

A = Matrix([
    [1.7, -0.6,  0.9],
    [1.8, -0.9, -0.4],
    [2.4, -0.6,  0.7]
])

B = Matrix([1.6, 1.3, 1.2])
```

Введем функцию метода Крамера:

```python
def kramer_method(A, B):
    # Считаем определитель матрицы A
    det_A = A.det()

    if det_A == 0:
        return 'Главный определитель равен ноль => корней нет'

    answer = []
    for i in range(A.shape[0]):
        Ai = A.copy()
        Ai[:, i] = B
        answer.append(Ai.det() / det_A)
    return answer
```

Для матрицы A вызовем функцию kramer_method() и получим ответ:

```python
x, y, z = kramer_method(A, B)
print(f'Методом Крамера получены следующие значения:\nX = {round(x, 3)}\nY = {round(y, 3)}\nZ = {round(z, 3)}')
```
