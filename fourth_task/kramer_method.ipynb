```python
from sympy import symbols, Matrix

A = Matrix([
    [1.7, -0.6,  0.9],
    [1.8, -0.9, -0.4],
    [2.4, -0.6,  0.7]
])

B = Matrix([1.6, 1.3, 1.2])
```

   :

```python
def kramer_method(A, B):
    #    A
    det_A = A.det()

    if det_A == 0:
        return '    =>  '

    answer = []
    for i in range(A.shape[0]):
        Ai = A.copy()
        Ai[:, i] = B
        answer.append(Ai.det() / det_A)
    return answer
```

  A   kramer_method()   :

```python
x, y, z = kramer_method(A, B)
print(f'    :\nX = {round(x, 3)}\nY = {round(y, 3)}\nZ = {round(z, 3)}')
```
