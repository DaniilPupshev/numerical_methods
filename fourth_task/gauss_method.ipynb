Запишем исходную матрицу (A|B):

```python
from sympy import pprint, Matrix

A_B = Matrix([
    [1.7, -0.6,  0.9, 1.6],
    [1.8, -0.9, -0.4, 1.3],
    [2.4, -0.6,  0.7, 1.2],
])
```

Далее приводим матрицу к ступенчатому виду:

```python
step_matrix = A_B.rref()[0]
pprint(step_matrix)
```

Из матрицы видно, что значения x, y, z будут равны:

```python
answer = []
for i in range(3, step_matrix.shape[0] * step_matrix.shape[1], 4):
    answer.append(step_matrix[i])

x, y, z = answer

# Выводим ответ
print(f'Методом Гаусса получены следующие значения:\nX = {round(x, 3)}\nY = {round(y, 3)}\nZ = {round(z, 3)}')
```
