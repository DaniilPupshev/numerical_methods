```python
import math
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

E = 10 ** (-4)
x = sp.symbols('x')

f = lambda x: (x - 1) * (2 ** x) - 1
f_derivative = sp.diff((x - 1) * (2 ** x) - 1, x) # первая производная - f'(x)
print(f"f'(x)={f_derivative}")
```

```python
f_d = lambda x: (2 ** x) * (x - 1) * math.log(2) + 2 ** x

# ==== Строим график f'(x), чтобы графически оценить max|f'(x)| на промежутке x ∈ [1;2] ====
array_x = np.linspace(1, 2, 1000)
y = [f_d(x_i) for x_i in array_x]

plt.plot(array_x, y, label="f'(x)=2^x*(x - 1)*log(2)+2^x")
plt.legend()
plt.grid(True)
plt.show()
```

Исходя из полученного графика выше можно сделать вывод, что max|f'(x)| на промежутке x ∈ [1;2] будет в точке x=2

```python
Q = abs(f_d(2))
print(f"f'(x) в точке x=2: {abs(f_d(2))}")
```

Теперь необходимо найти функцию fi(x) и параметр k:

```python
k = math.floor(Q / 2) # берем со знаком "+", т.к. функция на промежутке x ∈ [1;2] возрастает

# Окргуляем в большую сторону, чтобы k был целочисленный
while k % 2 != 0:
    k += 1

fi = lambda x: x - (f(x) / k)

print(f'fi(x) = x - (f(x) / k)\nk = {k}')
```
