```python
import math
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp

E = 10 ** (-4)
x = sp.symbols('x')

f = lambda x: (x - 1) * (2 ** x) - 1
f_derivative = sp.diff((x - 1) * (2 ** x) - 1, x) #   - f'(x)
print(f"f'(x)={f_derivative}")
```

```python
f_d = lambda x: (2 ** x) * (x - 1) * math.log(2) + 2 ** x

# ====   f'(x),    max|f'(x)|   x  [1;2] ====
array_x = np.linspace(1, 2, 1000)
y = [f_d(x_i) for x_i in array_x]

plt.plot(array_x, y, label="f'(x)=2^x*(x - 1)*log(2)+2^x")
plt.legend()
plt.grid(True)
plt.show()
```

       ,  max|f'(x)|   x  [1;2]    x=2

```python
Q = abs(f_d(2))
print(f"f'(x)   x=2: {abs(f_d(2))}")
```

    fi(x)   k:

```python
k = math.floor(Q / 2) #    "+", ..    x  [1;2] 

#    ,  k  
while k % 2 != 0:
    k += 1

fi = lambda x: x - (f(x) / k)

print(f'fi(x) = x - (f(x) / k)\nk = {k}')
```
