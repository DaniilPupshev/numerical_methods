```python
from math import cos, exp
import matplotlib.pyplot as plt

# Исходные данные
a, b = 0.0, 0.8
y = 1.1
h = 0.2     # шаг, подходящий по точности 10^(-4)

# Функция из уравнения
def f(x, y):
    return y * cos(x) - exp(-2 * x) * (2 + 3 * cos(x)) / (3 * y)

# Списки для хранения значений
x_values = [a]
y_values = [y]

# Метод Рунге–Кутты 4-го порядка
x = a
while x < b - 1e-10:
    k1 = f(x, y)
    k2 = f(x + h / 2, y + h * k1 / 2)
    k3 = f(x + h / 2, y + h * k2 / 2)
    k4 = f(x + h,   y + h * k3)
    y += h * (k1 + 2 * k2 + 2 * k3 + k4) / 6
    x += h
    x_values.append(round(x, 2))
    y_values.append(y)

# Вывод результатов
for i in range(len(x_values)):
    print(f"x = {x_values[i]:.2f}, y = {y_values[i]:.6f}")

# Построение интегральной кривой
plt.plot(x_values, y_values, marker='o')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Интегральная кривая y(x) методом Рунге–Кутта 4-го порядка")
plt.grid(True)
plt.show()
```
